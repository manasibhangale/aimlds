```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("basic line plot")
plt.show()
```

```python
x1=[1,2,3,4]
y1=[4,5,6,7]
plt.plot(x1,y1,color="red",linestyle="-.",marker="o",linewidth=2,markersize=9)
plt.grid(True)
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]

y1 = [1,2,3,4,5]
y2 = [2,4,6,8,10]
y3 = [1,4,9,16,25]
y4 = [5,4,3,2,1]

# Figure size
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("Plot 4")

plt.tight_layout()
plt.show()
```
