```python
import pandas as pd
import requests

url1 = "https://en.wikipedia.org/wiki/Mobile_country_code"

# Add browser headers
headers = {
    "User-Agent": "Mozilla/5.0"
}

# Fetch webpage
response = requests.get(url1, headers=headers)

# Read tables from HTML
df = pd.read_html(response.text, match="Country", header=0)[0]

print(df)
```
