```python
df['Marks']=df["Marks"].fillna(df["Marks"].mean())
```

```python
df.dtypes
```

```python
df=df.rename(columns={'Age':'Agee'})
df.head()
```

```python
df["Attendance_Percentage"]=df["Attendance_Percentage"].astype(float)
df.head(5)
```

```python
df["Semester"]=df["Semester"].apply(lambda x:x*2)
df.head(4)
```

```python
grouped_mean=df.groupby("Course")["Marks"].mean()
print(grouped_mean)
```

```python
grouped_sum=df.groupby(["Course","City"])["Marks"].sum()
print(grouped_sum)
```

```python
grouped_aggregate=df.groupby("City")["Marks"].agg(["mean","sum","count"])
print(grouped_aggregate)
```
