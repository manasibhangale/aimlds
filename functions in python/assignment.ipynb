```python

# convert C to F and F to C
def convert_temp(magnitude, unit):
    if unit.lower() == 'c':
        magnitude = (magnitude * 9 / 5) + 32
        print(f"{magnitude}F")
    elif unit.lower() == 'f':
        magnitude = (magnitude - 32) * 5 / 9
        print(f"{magnitude}C")

convert_temp(32, 'c')
```

```python
#password strength checker
def password_strength(password):
    if len(password)<8:
        print("not a strong password.")
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any (char in "!@#$%^&*()_+=-" for char in password):
        return False
    return True
password_strength("Bnangale@123")
```

```python
#calculate the total cost of items in shopping cart
def totalcost(cart):
    total_cost=0
    for item in cart:
        total_cost+=item["quantity"]*item["price"]
    return total_cost

cart=[
    {"item":"apple","price":50,"quantity":4},
    {"item":"mango","price":100,"quantity":4},
    {"item":"guava","price":25,"quantity":4}
]
total_cost=totalcost(cart)
print(total_cost)
```

```python
#is palindrome
def is_palindrome(s):
    s = s.lower().replace(" ", "")   
    reverse = s[::-1]
    if s == reverse:
        return True
    return False

print(is_palindrome("car"))
print(is_palindrome("racecar"))
```

```python
#calculate factorial using recursion
def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)
factorial(5)
```

```python
#function to read a file and count the frequency of each word
def count_freq(filepath):
    word_count={}
    with open(filepath,'r') as file:
        for line in file:
            words=line.split()
            for word in words:
                word=word.lower().strip("!@#$%^&*()_+?><,./';][=-]")
                word_count[word]=word_count.get(word,0)+1
    return word_count
filepath="manasi.txt"
count_freq(filepath)
```
