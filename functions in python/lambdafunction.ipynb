A lambda function is an inline function without a name, which can take any number of arguments but contains only a single expression, and automatically returns the result of that expression.
lambda arguments: expression
arguments → input parameters (like normal functions)
expression → a single computation (no statements like loops, assignments)

```python
#add two numbers
addition=lambda a,b:a+b
print(type(addition))
addition(3,2)
```

```python
#square of two numbers
square=lambda a:a**2
square(21)
```

```python
even=lambda a:a%2==0
even(9)
even(8)
```

```python
#largest of 3
(lambda a, b, c: a if a > b and a > c else (b if b > c else c))(3, 7, 5)
```

```python
#palindrome
(lambda s:s==s[::-1])("racecar")
```
