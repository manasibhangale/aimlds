## 🔹 Target Guided Ordinal Encoding (a.k.a. Mean Encoding)

**Target Guided Ordinal Encoding** is a feature encoding technique where categorical values are replaced with numbers based on their **relationship with the target variable**.

Instead of assigning arbitrary numbers (like label encoding), we assign values based on **how much each category influences the target**.

---

## 🔹 Core idea

For each category:

> Compute a statistic (usually **mean of target**) → use it as the encoded value

So categories are ordered by their **impact on the target variable**.

---

## 🔹 Example (Classification case)

Suppose we have:

| City | Target (Loan Approved: 1/0) |
| ---- | --------------------------- |
| A    | 1                           |
| A    | 1                           |
| A    | 0                           |
| B    | 0                           |
| B    | 0                           |
| C    | 1                           |
| C    | 0                           |

### Step 1: Compute mean of target per category

| City | Mean Approval Rate |
| ---- | ------------------ |
| A    | (1+1+0)/3 = 0.67   |
| B    | (0+0)/2 = 0.00     |
| C    | (1+0)/2 = 0.50     |

---

### Step 2: Replace categories with mean

| City | Encoded Value |
| ---- | ------------- |
| A    | 0.67          |
| B    | 0.00          |
| C    | 0.50          |

---

## 🔹 Python implementation

```python
import pandas as pd

df = pd.DataFrame({
    'City': ['A','A','A','B','B','C','C'],
    'Target': [1,1,0,0,0,1,0]
})

# Mean encoding
mean_encoded = df.groupby('City')['Target'].mean()

df['City_encoded'] = df['City'].map(mean_encoded)

print(df)
```

---

## 🔹 Why it is powerful

Used heavily in:

* Machine Learning competitions (Kaggle)
* Data Science tabular modeling

Because it:

* Captures **relationship with target**
* Works well for **high-cardinality categorical variables**
* Reduces dimensionality (unlike one-hot encoding)

---

## 🔹 Advantages

* Very **informative encoding**
* Works well with tree-based models
* Handles large number of categories efficiently

---

## 🔹 Major drawback ⚠️

It can cause **data leakage** because it uses the target variable.

---

## 🔹 How to fix leakage (important in interviews)

Use:

* **K-fold target encoding**
* **Leave-one-out encoding**
* **Smoothing (Bayesian encoding)**

These prevent overfitting by not using full target information directly.

---

## 🔹 Quick comparison

| Encoding Type          | Uses target? | Risk of leakage      | Best for                  |
| ---------------------- | ------------ | -------------------- | ------------------------- |
| Label Encoding         | ❌            | No                   | Tree models               |
| One-Hot Encoding       | ❌            | No                   | Nominal data              |
| Ordinal Encoding       | ❌            | No                   | Ordered data              |
| Target Guided Encoding | ✅            | Yes (if not careful) | High-cardinality features |
