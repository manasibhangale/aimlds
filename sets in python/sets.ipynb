```python
myset={1,2,3,4,5,5,5}
print(myset)
```

```python
#removes a particular element from the set
myset.remove(4)
print(myset)
```

```python
#removing an element that is not present in the set throws and error
myset.remove(11)
print(myset)
```

```python
#when we try to remove an element which is not present in the set it does not throw the error
myset.discard(99)
print(myset)
```

```python
#pop follows FIFO
popped=myset.pop()
print(myset)
```

```python
#clear all elements
myset.clear()
print(myset)
```

```python
#mathematical functions
set1={1,2,3,4,5}
set2={4,5,6,7,8}
#union
print(set1.union(set2))
#intersection
print(set1.intersection(set2))
#intersectionupdate
set1.intersection_update(set2)
print(set1)
```

```python
set1={1,2,3,4,5}
set2={4,5,6,7,8}
print(set1.issubset(set2))
print(set1.issuperset(set2))
print(set1.isdisjoint(set2))
```

```python
print(set1.difference(set2))
print(set1.symmetric_difference(set2))
print(set2.difference(set1))
```
