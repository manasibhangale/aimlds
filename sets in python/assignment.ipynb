```python
#create set with first 10 positive integers
set1=set(range(1,11))
print(set1)
```

```python
#add 11 to the set 1 and then remove 1
set1.add(11)
print(set1)
set1.remove(1)
print(set1)
```

```python
set1=set(range(1,6))
set2=set(range(2,11,2))
print(set1) 
print(set2)
set1.union(set2)
set1.intersection(set2)
set1.difference(set2)
set1.symmetric_difference(set2)
```

```python
squares=set(x**2 for x in range(1,11))
print(squares)
```

```python
#print only the evens from the set1 using set comprehension
set1=set(range(1,11))
print(set1)
even=set(x for x in set1 if x%2==0)
print(even)
```

```python
set2={1,1,1,2,2,2,3,3,3,4,4,5,55}
unique=set(set2)
print(unique)
```

```python
set1=set(range(1,6))
set2=set(range(1,4))
print(set2.issubset(set1))
print(set1.issuperset(set2))
```

```python
fs = frozenset(range(1, 6))
print(fs)
```

```python
set1=set(range(1,7))
list1=list(set1)
list1.append(6)
print(set(list1))
```
