Magic Methods in Python

Magic methods are special methods in Python that start and end with double underscores (__).

They are also called:

Dunder methods (Double Under methods)
Special methods

These methods allow objects to behave like built-in Python objects.
Why are Magic Methods Used?

They are used to:

Initialize objects
Define object representation
Perform operator overloading
Compare objects
Make objects iterable
Support indexing, length, etc.

```python
class Student:
    def __init__(self, name):
        self.name = name

s = Student("Manasi")
print(s.name)
```

```python
class Student:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Student name is {self.name}"

s = Student("Manasi")
print(s)
```

```python
class Books:
    def __init__(self, books):
        self.books = books

    def __len__(self):
        return len(self.books)

b = Books(["Python", "Java", "C++"])

print(len(b))
```

```python
class Number:
    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return self.value + other.value

n1 = Number(10)
n2 = Number(20)

print(n1 + n2)
```

```python
class Number:
    def __init__(self, value):
        self.value = value

    def __sub__(self, other):
        return self.value - other.value

n1 = Number(50)
n2 = Number(20)

print(n1 - n2)
```

```python
class Number:
    def __init__(self, value):
        self.value = value

    def __mul__(self, other):
        return self.value * other.value

n1 = Number(5)
n2 = Number(4)

print(n1 * n2)
```

```python
class Number:
    def __init__(self, value):
        self.value = value

    def __truediv__(self, other):
        return self.value / other.value

n1 = Number(20)
n2 = Number(5)

print(n1 / n2)
```

```python
class Student:
    def __init__(self, marks):
        self.marks = marks

    def __eq__(self, other):
        return self.marks == other.marks

s1 = Student(90)
s2 = Student(90)

print(s1 == s2)
```

```python
class Number:
    def __init__(self, value):
        self.value = value

    def __lt__(self, other):
        return self.value < other.value

n1 = Number(10)
n2 = Number(20)

print(n1 < n2)
```

```python
class MyList:
    def __init__(self, items):
        self.items = items

    def __getitem__(self, index):
        return self.items[index]

obj = MyList([10, 20, 30])

print(obj[1])
```

```python
class Demo:
    def __call__(self):
        print("Object called like function")

d = Demo()

d()
```

```python
class Demo:
    def __del__(self):
        print("Object destroyed")

d = Demo()
```

| Magic Method  | Purpose               |
| ------------- | --------------------- |
| `__init__`    | Constructor           |
| `__str__`     | String representation |
| `__len__`     | Length                |
| `__add__`     | `+` operator          |
| `__sub__`     | `-` operator          |
| `__mul__`     | `*` operator          |
| `__truediv__` | `/` operator          |
| `__eq__`      | `==` operator         |
| `__lt__`      | `<` operator          |
| `__getitem__` | Indexing              |
| `__call__`    | Callable object       |
| `__del__`     | Destructor            |
